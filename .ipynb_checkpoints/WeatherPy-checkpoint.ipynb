### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.
* There is no strong relationship between latitude and cloudiness. However, it is interesting to see that a strong band of cities sits at 0, 80, and 100% cloudiness.
* There is no strong relationship between latitude and wind speed. However, in northern hemispheres there is a flurry of cities with over 20 mph of wind.

```python
# Dependencies and Setup
# %matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print("Beginning Data Retrieval")
print("-------------------------")

RecordCount=0
SetCount=1
row=0

CitiesDF =pd.DataFrame()

CitiesDF["City"] =""
CitiesDF["Cloudiness"] =""
CitiesDF["Country"] =""
CitiesDF["Date"] =""
CitiesDF["Humidity"] =""
CitiesDF["Lat"] =""
CitiesDF["Lng"] =""
CitiesDF["Max Temp"] =""
CitiesDF["Wind Speed"] =""



for city in cities: 
    
    if RecordCount==50 :
        SetCount+=1
        RecordCount=0 
          
    
    print(f"Processing Record {RecordCount} of Set {SetCount} | {city}")
    
    query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={api_key}&q={city}&units=imperial"
    try:
        weather_response = requests.get(query_url).json()
        CitiesDF.loc[row,"City"] = weather_response['name']
        CitiesDF.loc[row,"Cloudiness"] = weather_response['clouds']['all']
        CitiesDF.loc[row,"Country"] =weather_response['sys']['country']
        CitiesDF.loc[row,"Date"] =weather_response['dt']
        CitiesDF.loc[row,"Humidity"] =weather_response['main']['humidity']
        CitiesDF.loc[row,"Lat"] =weather_response['coord']['lat']
        CitiesDF.loc[row,"Lng"] =weather_response['coord']['lon']
        CitiesDF.loc[row,"Max Temp"] =weather_response['main']['temp_max']
        CitiesDF.loc[row,"Wind Speed"] =weather_response['wind']['speed']
    except KeyError:
        print("City not found. Skipping...")
    row +=1   
    RecordCount+=1 
print("-----------------------------")
print("Data Retrieval Complete")      
print("-----------------------------")  
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
CitiesDF.to_csv("WeatherDetailsForCitiesData.csv")

print(CitiesDF.count())

CitiesDF.head()
```

#### Latitude vs. Temperature Plot

```python
x_axis = CitiesDF["Lat"]
y_axis = CitiesDF["Max Temp"]

plt.scatter(x_axis, y_axis, marker="o", facecolors="#0086b3", edgecolors="black", alpha=0.7)
plt.xlabel("Latitude")
plt.ylabel("Max Temp(F)")
CurrentDate = time.strftime("%m/%d/%y")
plt.title("City Latitude vs. Max Temperature (" + CurrentDate + ")")
plt.grid()
plt.savefig("Latitude vs. Temperature Plot.png")
```

#### Latitude vs. Humidity Plot

```python
x_axis = CitiesDF["Lat"]
y_axis = CitiesDF["Humidity"]

plt.scatter(x_axis, y_axis, marker="o", facecolors="#0086b3", edgecolors="black", alpha=0.7)
plt.xlabel("Latitude")
plt.ylabel("Humidity %")
CurrentDate = time.strftime("%m/%d/%y")
plt.title("City Latitude vs. Humidity (" + CurrentDate + ")")
plt.ylim(0,105)
plt.grid()
plt.savefig("Latitude vs. Humidity Plot.png")
```

#### Latitude vs. Cloudiness Plot

```python
x_axis = CitiesDF["Lat"]
y_axis = CitiesDF["Cloudiness"]

plt.scatter(x_axis, y_axis, marker="o", facecolors="#0086b3", edgecolors="black", alpha=0.7)
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")
CurrentDate = time.strftime("%m/%d/%y")
plt.title("City Latitude vs. Cloudiness(" + CurrentDate + ")")
plt.grid()
plt.savefig("Latitude vs. Cloudiness.png")
```

#### Latitude vs. Wind Speed Plot

```python
x_axis = CitiesDF["Lat"]
y_axis = CitiesDF["Wind Speed"]

plt.scatter(x_axis, y_axis, marker="o", facecolors="#0086b3", edgecolors="black", alpha=0.7)
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
CurrentDate = time.strftime("%m/%d/%y")
plt.title("City Latitude vs. Wind Speed (" + CurrentDate + ")")
plt.ylim(-1, 45)
plt.grid()
plt.savefig("Latitude vs. Wind Speed.png")
```
